```python
#apply logistic regressions
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df=pd.get_dummies(df)
```

```python
x=df.drop(['Business_Sourced'],axis=1)
y=df['Business_Sourced']
```

```python
ss.fit_transform(x,y)
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(x,y,stratify=y,test_size=0.3)
train_x.shape,test_x.shape
```

## random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(max_depth=10,min_samples_split=8)
```

```python
rfc.fit(train_x,train_y)
```

```python
train=rfc.predict_proba(train_x) 
```

```python
test=rfc.predict_proba(test_x)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(train_y,train[:,1])
```

```python
roc_auc_score(test_y,test[:,1])
```

- Here we can see that when we do not remove the PIN no var it incsresed the accuracy(auc-roc) of model.
- we can set different paramters for RFC an dsee the change in values like when max depth=10 it get best roc-auc score
